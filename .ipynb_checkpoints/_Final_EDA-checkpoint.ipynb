# Rossmann Store Sales Prediction using Machine Learning

## Project Overview

Rossmann operates over 1,000 retail stores across several European countries. Accurate sales forecasting enables better inventory planning, workforce management, promotional planning, and business decision-making.

This project aims to analyze historical sales data from Rossmann stores and build a machine learning model capable of predicting future daily sales based on store characteristics, promotional activities, competition information, and temporal features.

---

# Business Problem

Retail businesses require accurate sales forecasts to optimize inventory, staffing, and promotional strategies.

The objective is to predict the **daily sales** of Rossmann stores using historical transactional and store information while identifying the most important factors influencing sales.

---

# Project Objectives

- Perform comprehensive Exploratory Data Analysis (EDA).
- Understand customer purchasing behavior.
- Analyze the impact of promotions, holidays, competition, and seasonality.
- Engineer meaningful features from raw data.
- Build and compare multiple machine learning regression models.
- Select the best-performing model for sales forecasting.

---

# Dataset Information

The project uses the **Rossmann Store Sales** dataset, which consists of historical daily sales records and store metadata.

### Main Files

- **train.csv** – Historical daily sales data.
- **store.csv** – Store-related information such as store type, assortment, competition, and promotional details.
- **test.csv** – Future records for which sales need to be predicted.

---

# Target Variable

**Sales**

The target variable represents the total daily sales generated by each Rossmann store.

---

# Features Used

The dataset contains features related to:

- Store Characteristics
- Promotional Activities
- Competition Information
- Holiday Information
- Customer Footfall
- Temporal Features (Year, Month, Quarter, Day, DayOfWeek, Season)

Several additional features such as **CompetitionDuration**, **PromoDuration**, **Season**, and **IsPromoMonth** were engineered during feature engineering to improve predictive performance.

```python
df["DistanceGroup"] = pd.cut(
    df["CompetitionDistance"],
    bins=[0,1000,5000,10000,20000,80000],
    labels=[
        "0-1 km",
        "1-5 km",
        "5-10 km",
        "10-20 km",
        ">20 km"
    ]
)
```

```python
df.groupby('DistanceGroup')[['Sales']].mean()
```

### Average sales vary across different competition distance groups, but no clear increasing or decreasing trend is observed. This suggests that CompetitionDistance alone is not a strong determinant of sales, and other factors such as promotions, assortment, store type, and customer count also influence sales.

### Why is Promo2 opposite to Promo?

- This is expected.

- Promo (short-term offer) → immediately attracts customers → Sales increase.
- Promo2 (continuous promotion) → often used in stores that already have weaker demand, so despite the promotion, their average sales can still be lower.

- This is a common business insight

### Observations

- The dataset is almost evenly distributed between stores participating in Promo2 (49.87%) and those not participating (50.13%).
- Stores without Promo2 have a higher average sales value (7351) than stores with Promo2 (6558).
- This suggests that continuous promotions (Promo2) do not necessarily lead to higher sales and may be implemented in stores with relatively lower baseline demand.
- Across all store types, average sales remain lower for Promo2 stores, although Store Type B continues to record the highest sales in both groups.
- The observed relationship indicates that Promo2 interacts with Store Type and should be retained as a predictive feature for the sales forecasting model.

## 7.PromoInterval

```python
df['PromoInterval'].info()
```

```python
df.groupby("Store")["PromoInterval"].nunique().loc[lambda x: x > 1]
```

```python
month_map = {
    1: "Jan",
    2: "Feb",
    3: "Mar",
    4: "Apr",
    5: "May",
    6: "Jun",
    7: "Jul",
    8: "Aug",
    9: "Sept",
    10: "Oct",
    11: "Nov",
    12: "Dec"
}

def is_promo_month(month, promo2, interval):

    if promo2 == 0:
        return 0

    month_name = month_map[month]

    if month_name in interval:
        return 1
    else:
        return 0

df["IsPromoMonth"] = df.apply(
    lambda x: is_promo_month(
        x["Month"],
        x["Promo2"],
        x["PromoInterval"]
    ),
    axis=1
)
```

# Final EDA Summary

## Conclusion

The Exploratory Data Analysis (EDA) was conducted on the Rossmann Store Sales dataset to understand the key factors affecting daily store sales and customer behavior. The analysis covered target variable exploration, numerical features, categorical features, engineered features, and time-based features. Several important business insights were identified, which will guide feature selection and machine learning model development.

---

# Numerical Feature Analysis

- The target variable **Sales** exhibits a moderately right-skewed distribution, indicating that most stores generate moderate sales while a relatively small number of stores achieve exceptionally high sales.
- **Customers** has the strongest positive relationship with Sales, confirming that customer footfall is the primary driver of store revenue.
- **CompetitionDistance** has only a weak influence on Sales, suggesting that the distance to the nearest competitor alone is not a strong determinant of store performance.
- **CompetitionDuration** shows a mild positive relationship with Sales. Stores that have coexisted with competitors for longer periods tend to exhibit slightly more stable sales patterns.
- **PromoDuration** captures the duration of recurring promotional campaigns and is expected to provide useful information during model training.
- Correlation analysis indicates that **Customers**, **Promo**, and **Open** (before removing closed stores) have the strongest relationship with Sales, while several other features contribute moderately.

---

# Categorical Feature Analysis

- **StoreType** significantly influences Sales, with different store formats generating noticeably different average revenue.
- **Assortment** has a clear impact on store performance. Stores offering a wider assortment generally achieve higher average sales.
- **Promo** is one of the strongest business drivers, substantially increasing both customer traffic and daily sales.
- **Promo2** represents long-term recurring promotional campaigns and behaves differently from the regular Promo feature.
- **StateHoliday** affects sales because customer purchasing behavior and store operations vary during public holidays.
- **SchoolHoliday** shows only a moderate impact on Sales.
- **PromoInterval** reveals the recurring promotional schedule followed by each store.
- A new engineered feature, **IsPromoMonth**, was created to identify whether a store is currently in an active recurring promotion month, making promotional information easier for the model to learn.
- **Season** also influences purchasing behavior, demonstrating clear seasonal variations in sales.

---

# Time-Based Feature Analysis

- Sales show a gradual improvement from **2013 to 2015**, indicating overall business growth.
- **Month** exhibits strong seasonality, with sales increasing significantly during **November and December**, reflecting holiday shopping behavior.
- **Quarter** analysis indicates that **Q4** consistently records the highest sales and customer traffic.
- **Day of Week** has a strong impact on sales. Sales vary considerably across weekdays, making it an important predictor.
- **Day of Month** reveals cyclical purchasing behavior, with higher sales generally occurring at the beginning and end of each month.
- Overall, temporal features capture significant seasonal and calendar-based purchasing patterns.

---

# Business Insights

- Customer footfall is the strongest driver of store sales.
- Promotional campaigns significantly increase both customer traffic and revenue.
- Store characteristics such as **StoreType** and **Assortment** strongly influence sales performance.
- Sales exhibit clear seasonal patterns across months, quarters, weekdays, and seasons.
- **December** consistently records the highest average sales due to holiday demand.
- Beginning-of-month and end-of-month shopping patterns indicate possible salary-cycle purchasing behavior.
- Competition-related variables influence sales, but their impact is considerably smaller than promotional and customer-related factors.
- Engineered features such as **CompetitionDuration**, **PromoDuration**, and **IsPromoMonth** capture additional business information that may improve predictive performance.

---

# Important Modeling Consideration

During EDA, **Customers** was identified as the feature most strongly correlated with Sales.

However, the objective of this project is to **forecast future sales**, and future customer counts are not available at prediction time. Therefore, using the **Customers** feature during model training would introduce **target leakage**, resulting in unrealistic model performance.

Although Customers provides valuable business insights during EDA, it **will not be used as an input feature** for the final forecasting model.

Instead, the model will rely only on features that are available before the prediction date, including promotional information, store characteristics, competition-related features, and time-based features.

The following features will not be used during model development.

| Date | Already decomposed into Year, Month, Quarter, Day, DayOfWeek, and Season. |
| Open | Contains only one unique value after removing closed-store records. |
