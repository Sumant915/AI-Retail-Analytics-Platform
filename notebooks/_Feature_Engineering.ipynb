### CompetitionDuration Feature

- Created a new feature named `CompetitionDuration` by calculating the number of months between the transaction date and the competition opening date.
- Negative duration values were replaced with `0`, indicating that the competitor had not yet opened at the time of the transaction.
- Missing values in `CompetitionDuration` were imputed using the median because the feature exhibited a positively skewed distribution with a few extreme values.
- This engineered feature provides a better representation of the competitive environment than using separate competition month and year columns.

# Feature Engineering Summary

### Date-based Features
- Extracted `Year`, `Month`, `Day`, and `Quarter` from the `Date` column to capture yearly, monthly, daily, and quarterly sales trends.
- Created a new `Season` feature by grouping months into Winter, Spring, Summer, and Autumn. This feature helps capture seasonal demand patterns that may influence store sales.

### CompetitionDuration
- Created a new feature `CompetitionDuration` by calculating the number of months between the transaction date and the competition opening date.
- This feature represents how long a nearby competitor had been operating at the time of each transaction.
- Negative values were replaced with `0`, indicating that the competitor had not yet opened during the transaction period.
- Missing values were imputed using the median since the competition opening date was unavailable for some stores.

### PromoDuration
- Converted `Promo2SinceYear` and `Promo2SinceWeek` into an actual calendar date (`PromoStartDate`) using the ISO week format.
- Created a new feature `PromoDuration` by calculating the number of months between the transaction date and the promotion start date.
- This feature represents the duration for which a continuous promotion (Promo2) had been running at the time of each transaction.
- Negative values were replaced with `0`, indicating that the promotion had not yet started.
- Stores with `Promo2 = 0` were assigned a `PromoDuration` of `0`, as they were not participating in the continuous promotion program.

### Removed Features
- Removed `CompetitionOpenSinceMonth` and `CompetitionOpenSinceYear` after creating the `CompetitionDuration` feature, as the new feature captures the same information in a more meaningful form.
- Removed `Promo2SinceWeek`, `Promo2SinceYear`, and the temporary `PromoStartDate` column after creating the `PromoDuration` feature, as these columns were no longer required.

### Overall Observation
- Feature engineering transformed raw date, competition, and promotion information into meaningful business-oriented features.
- The newly created features better represent seasonality, competition history, and promotion history, which are expected to improve the performance and interpretability of the sales forecasting model.
- CompetitionDuration: How long the competitor has been operating near the store at the time of sale.
- PromoDuration: How long the continuous promotion has been running at the time of sale.
